```python
def reverse():
    n=int(input("Enter Number"))    #n = 121,12, 1,0
    rev=0                           #rev = 0,1,12,121
    while n!=0:
        d=n%10     #returns reminder     #d = 1,2,1
        rev=rev*10+d   #reverse formula    rev = 1,12,121
        n=n//10       #//  gives only integer and ignores decima point   n = 12,1,0
    #print(rev)
    return rev
```

```python
#write a program which will check if given number is palindrome or not
def palindrome():
  n = int(input("Enter Number"))
  temp=n
  rev = 0
  while n!=0:
    d = n%10
    rev = rev*10+d
    n=n//10
  if rev==temp:
    print("Number is palindrome")
  else:
    print("number is not palindrome")

palindrome()
```

#2. Keyword based argument.

```python
def greek(name,msg):
  print(name,msg)
```
