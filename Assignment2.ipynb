```python
#1.	Write a function to check whether a number is prime.
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input())
if is_prime(num):
    print("Prime")
else:
    print("Not Prime")
```

```python
#2.Write a function to calculate factorial of a number.
```

```python
def factorial(n):
    if n < 0:
        return "Not defined"
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

num = int(input())
print(factorial(num))
```

```python
# 4. try-except-else
try:
    a = int(input())
    b = int(input())
    result = a / b
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
else:
    print(result)
```

```python
#5.	Create a custom exception to validate age greater than 18.
```

```python
# 5. Custom exception for age > 18
class AgeError(Exception):
    pass

try:
    age = int(input())
    if age <= 18:
        raise AgeError("Age must be greater than 18")
    print("Valid age")
except AgeError as e:
    print(e)
except ValueError:
    print("Invalid input")
```
