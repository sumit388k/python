```python
a=int(input("enter marks"))
if a <= 39 :
  print("you are fail")
elif a >= 40 and a <= 50 :
  print("you are pass")
elif a >= 51 and a <= 60 :
   print("you are in a second class")
elif a >= 61 and a <= 65 :
  print("you are in a first class")
elif a >= 66 and a <= 100 :
  print("you are in a distinction")
```
