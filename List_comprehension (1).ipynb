```python
#WAP to squre each element of list
a = [2,3,4,5,6]    #[4,9,14,25,36]
result = []
for i in a:
  result.append(i*i)
print(result)
```

```python
result = [i*i for i in a]   #using list comprehension
print(result)
```

```python
a = [i**2 for i in range(1,11)]    #range(1,11)---> to to 10 integers in list
```

```python
a = [i for i in range(1,11) if i%2==0]
print(a)
```

```python
#WAP to square odd numbers from given list
result=[]
a = [1,2,3,4,5,6,7,8]
for i in a:
  if i%2!=0:
    result.append(i*i)
print(result)
```

```python
res = [i*i for i in a if i%2!=0]
res
```
