```python
class Student:
  """This is a student class"""
  def __init__(self,name,rollno,marks):             ##__init__(self) it is called constructor. COnstructor is used to initalise the object. WHenever object gets created it will invoke
    self.name=name
    self.rollno=rollno
    self.marks=marks
  def display(self):
      print("Student Name:",self.name)
      print("Student Rollno:",self.rollno)
      print("Student Marks:",self.marks)


e = Student(name="amit",rollno=1,marks=87)
e.display()
e1=Student(name="abc",rollno=2,marks=90)
e1.display()
```

```python
class Test:
    x=10  #static /class variable
    def __init__(self):
        self.a=20
        Test.z=1000
    def Demo(self):
        Test.c=109


t = Test()
print(Test.x)
print(t.a)
t.Demo()
print(dir(Test))
```

```python
#Create class customer
# bankname as static variable
# deposite()
# withdraw() You can withdraw if bal>1000
# othert wise give error message.

# Accept bankdetails and display it with display()
# Perform deposite() and withdraw ()

# finally display final balance after each transaction
```

```python
class Customer:
  bankname = "ABI"
  def __init__(self,name,bal,acc):
    self.name = name
    self.bal = bal
    self.acc_no = acc
  def diposite(self):
    amount = int(input("enter amount to diposite"))
    self.bal = self.bal + amount
    print("amount diposited")
  def withdraw(self):
    amount = int(input("enter amount to withdraw"))
    if self.bal > 1000:
      self.bal = self.bal - amount
      print("amount withdraw")
    else:
      print("insufficient balance")
  def check_bal(self):
    print("available balance is",self.bal)


  def display(self):
    print("Name:",self.name)
    print("Account Number:",self.acc_no)
    print("Available Balance:",self.bal)


amit = Customer(input("Enter name:"),int(input("Enter balance:")),int(input("Enter account number:")))
amit.diposite()
amit.withdraw()
amit.check_bal()
amit.display()
```

```python
# inhertance
#Inheritance is a core concept in Object-Oriented Programming (OOP) that allows a child (derived) class to acquire the properties and behaviors (methods) of a parent (base) class.
```

```python
class Parent:
  def m1(self):
    print("Parent method")

class Child(Parent):
  def m2(self):
    print("Child method")


c = Child()
print(dir(c))
c.m1()
c.m2()
```

```python
#2. Multilevel inheritance.
#In this one parent may have one clild and that child may have other child so here parent will act as grandparent for last chlid.
class P:
    def m1(self):
        print("This is parent class method")
class B(P):
    def m2(self):
        print("This is child class of P")
class C(B):
    def m3(self):
        print("This is child of B and grand child of A")

c = B()
dir(c)
```
