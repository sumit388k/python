# **Implicit Type Casting (Automatic)**

```python
a = 5        # int
b = 2.5      # float

c = a + b    # int + float = float
print(c)     # 7.5
print(type(c))  # float
```

# Explicit Type Casting (Manual)

```python
x = "100"                 #String → Integer
y = int(x)
print(y, type(y))
```

```python
a = 5.89              #Float → Integer (decimal removed)
b = int(a)
print(b)   # 5
```

```python
#Tuple → List
t = (1, 2, 3)
l = list(t)
print(l)
```

```python
#List → Set
lst = [1, 2, 3, 3]
st = set(lst)
print(st)   # {1,2,3}
```

```python
#Integer → String
num = 10
s = str(num)
print(s, type(s))
```
