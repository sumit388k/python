```python
# 6. Count digits
# Take an integer input and count the number of digits using a while loop.
```

```python
num = int(input("Enter an integer: "))

count = 0

if num == 0:
    count = 1
else:
    while num != 0:
        num = num // 10
        count += 1

print("Number of digits:", count)
```

```python
#19. Armstrong number
# Check if a number is an Armstrong number using loop and conditions.       Example: 153 â†’ 1Â³ + 5Â³ + 3Â³ = 153
```

```python
n = int(input("Enter a number: "))
temp = n
sum = 0

digits = len(str(n))

while temp > 0:
    digit = temp % 10
    sum += digit ** digits
    temp = temp // 10

if sum == n:
    print(n, "is an Armstrong number")
else:
    print(n, "is not an Armstrong number")
```

```python
#20. Palindrome number
# Check whether a number is palindrome using loops.    Example: 1221 â†’ palindrome, 123 â†’ not palindrome
```

```python
n = int(input("Enter a number: "))
temp = n
reverse = 0

while temp > 0:
    digit = temp % 10
    reverse = reverse * 10 + digit
    temp = temp // 10

if reverse == n:
    print(n, "is a palindrome number")
else:
    print(n, "is not a palindrome number")
```

```python
# 21. Find all prime numbers in a range
```

```python
start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))

print("Prime numbers in the range are:")

for num in range(start, end + 1):
    if num > 1:
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, end=" ")
```

```python
# 22. Input two numbers a and b.
# Print all prime numbers between them using loops + conditions.
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Prime numbers between", a, "and", b, "are:")

for num in range(a, b + 1):
    if num > 1:
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, end=" ")
2
```
